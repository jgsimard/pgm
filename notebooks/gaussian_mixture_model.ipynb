## GMM derivation

\begin{align*}
    Z_i &\overset{iid}{\sim} Mult(\pi)\\
    X_i|Z_i &= j \sim \mathcal{N}(\mu_j,\Sigma_j) \\
    \theta &= (\pi, \{\mu_j\}_{j=1}^k, \{\Sigma_j\}_{j=1}^k) \\
    l(\theta) &= \sum\limits_{i=1}^n [\log p(x_i|z_i; \theta) + \log p(z_i; \theta)] \\
    &= \sum\limits_{i=1}^n \left[\sum\limits_{j=1}^k z_{i,j} \log \mathcal{N}(x_i|\mu_j,\Sigma_i) + \sum\limits_{j=1}^k z_{i,j} \log(\pi_j) \right] \\
    &= \sum\limits_{i=1}^n \sum\limits_{j=1}^k z_{i,j} [\log \mathcal{N} (x_i|\mu_j,\Sigma_i) + \log(\pi_j)]
\end{align*}
